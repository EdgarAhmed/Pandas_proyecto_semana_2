```python
def limpiar_activ(x):
    
    if 'surf' in x:
        return 'Surfing'
    if 'swim' in x:
        return 'Swimming'
    if 'Fish' in x:
        return 'Fishing'
    if 'Spear' in x:
        return 'Spear Fishing'
    else:
        return 'Others'
```

```python
df[['Activity']].value_counts(normalize=True)*100
```

```python
df['Activity']=df['Activity'].apply(limpiar_activ)

df.head()
```

```python
df.Activity.isna().sum() #no hay vacios
```

```python
df[['Activity']].value_counts(normalize=True)*100
```

```python
import matplotlib.pyplot as plt

def pie_chart(data_series, title):
    # Crea el gráfico circular
    plt.figure(figsize=(8, 8))
    plt.pie(data_series, labels=data_series.index, autopct='%1.1f%%', startangle=140)
    # Configura el título
    plt.title(title)

    # Muestra el gráfico
    plt.show()
```

```python
n_activities= df['Activity'].value_counts(normalize=True)*100
```

```python
pie_chart(n_activities, 'pie')
```
