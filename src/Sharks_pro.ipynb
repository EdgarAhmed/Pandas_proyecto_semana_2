```python
df['Activity'].fillna('Unknown', inplace=True)  #Relleno los vacio con desconocidos
```

```python
df.Activity.isna().sum()  #comprobación de si hay vacias. 
```

```python
#función para quedarme con los principales tipos de actividades que produzcan ataques. 
#Surf, nadar, todo lo que tenga que ver con pesca y similares. El resto, entrará en la categoria 'otros'

def limpiar_activ(x):
    
    if 'surf' in x:
        return 'Surfing'
    if 'swim' in x:
        return 'Swimming'
    if 'Fish' in x:
        return 'Fishing'
    if 'Spear' in x:
        return 'Spear Fishing'
    else:
        return 'Others'
```

```python
df[['Activity']].value_counts(normalize=True)*100 #comprobación de los porcentajes, una vez más.
```

```python
df['Activity']=df['Activity'].apply(limpiar_activ)

df.head()
```

```python
df.Activity.isna().sum() #no hay vacios
```

```python
df[['Activity']].value_counts(normalize=True)*100 #una vez aplicada la función se ve como se distribuyen ahora las actividades. 
```

```python
#para crear un Pie Chart

import matplotlib.pyplot as plt

def pie_chart(data_series, title):
    # Crea el gráfico circular
    plt.figure(figsize=(8, 8))
    plt.pie(data_series, labels=data_series.index, autopct='%1.1f%%', startangle=140)
    # Configura el título
    plt.title(title)

    # Muestra el gráfico
    plt.show()
```

```python
n_activities= df['Activity'].value_counts(normalize=True)*100
```

```python
pie_chart(n_activities, 'Actividades donde más ataques de tiburón se sufre')
```

```python
#Decido quedarme con los 5 paises donde más ataques suceden, el resto, que ponga 'otros'

def limpiar_paises(x):

    
    if 'USA' in x:
        return 'USA'
    if 'AUSTRALIA' in x:
        return 'AUSTRALIA'
    if 'SOUTH AFRICA' in x:
        return 'SOUTH AFRICA'
    if 'PAPUA NEW GUINEA' in x:
        return 'PAPUA NEW GUINEA'
    if 'NEW ZEALAND' in x:
        return 'NEW ZEALAND'
    else:
        return 'Others'
```
