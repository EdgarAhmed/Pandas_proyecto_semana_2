```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ejemplo de DataFrame
data = {
    'Año': [2010, 2011, 2012, 2013, 2014],
    'Ventas': [100, 130, 160, 140, 180]
}

df = pd.DataFrame(data)

# Crear el gráfico de líneas
plt.figure(figsize=(8, 6))  # Tamaño del gráfico opcional

plt.plot(df['Año'], df['Ventas'], marker='o', linestyle='-', color='b', label='Ventas')

plt.title('Ventas a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Ventas')
plt.grid(True)
plt.legend()

# Mostrar el gráfico
plt.show()
```
